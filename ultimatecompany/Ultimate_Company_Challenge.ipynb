## Pre-Processing and Training Data Development

```python
# Checking datatype again
df.dtypes 
```

### Categorical Encoding

```python
df_cats = df.iloc[:,[0,6,8,12]]
```

```python
df_cats.head() 
```

```python
# Using Label Encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_cats.city = le.fit_transform(df_cats.city)
df_cats.phone = le.fit_transform(df_cats.phone)
df_cats.ultimate_black_user = le.fit_transform(df_cats.ultimate_black_user)
df_cats.active = le.fit_transform(df_cats.active)
```

```python
df_cats.head() 
```

### Transformation of Numerical Variables

```python
# Using log function to normalize or scale the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

col_transform = df.iloc[:,[1, 3, 4, 7, 9, 10, 11]].values

columns=['trips_in_first_30_days', 'avg_rating_of_driver', 'avg_surge', 'surge_pct', 'weekday_pct', 
         'avg_dist', 'avg_rating_by_driver']

col_scaled = scaler.fit_transform(col_transform)

df_scaled = pd.DataFrame(col_scaled, columns=['trips_in_first_30_days', 'avg_rating_of_driver', 
                                              'avg_surge', 'surge_pct', 'weekday_pct', 'avg_dist', 'avg_rating_by_driver'])
```

```python
# Combining both df_cats and df_scaled
df_ml = pd.concat([df_cats,df_scaled], axis=1)
df_ml.head()
```

```python
# Splitting into X and y
X = df_ml.drop('active', axis=1)
y = df_ml['active']
```

```python
# Splitting into training and testing datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)
```

## Machine Learning Modeling

### Logistic Regression

```python
# Importing required libraries and packages for logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)

y_pred_lr = model_lr.predict(X_test)

print("Accuracy Score for the Logistic Regression Model is: ", accuracy_score(y_test, y_pred_lr))
```

The accuracy score is not very encouraging so I am going to try now Random Forest Model.

### Random Forest

```python
# Importing Random Forest packages and libraries
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=300, random_state = 11, n_jobs=-1)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)

print("Accuracy Score for the Random Forest Model is: ", accuracy_score(y_test, y_pred_rf))
```

This is a very good improvement on Logistic Regression. So I am going to do hyperparameter tuning now for the Random Forest modeling.

### Random Forest Hpyerparameter Tuning

```python
# Importing required packages and libraries
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score
from sklearn.metrics import average_precision_score, f1_score, precision_score, plot_confusion_matrix
from sklearn.metrics import roc_curve

def model_result(clf,x_test,y_test):
    y_prob=clf.predict_proba(x_test)
    y_pred=clf.predict(x_test)
    print('AUPRC :', (average_precision_score(y_test, y_prob[:, 1])))
    print('F1-score :',(f1_score(y_test,y_pred)))
    print('Confusion_matrix: ')
    print(confusion_matrix(y_test,y_pred))
    print("accuracy_score")
    print(accuracy_score(y_test,y_pred))
    print("classification_report")
    print(classification_report(y_test,y_pred))
```
